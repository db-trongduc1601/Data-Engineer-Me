```python
# Tạo một cái hộp (biến) để đựng tên của cậu
ten_cua_toi = "Bui Trong Duc"

# Yêu cầu máy tính in cái tên đó ra màn hình cùng lời chào
print(f"Xin chào Data Engineering! Tôi là {ten_cua_toi}")
```

```python
# Tạo một danh sách (List) chứa độ đục của 5 mẫu nước
danh_sach_do_duc = [2, 7, 3, 9, 1]

# Chạy băng chuyền (Vòng lặp For) để xét từng mẫu nước trong danh sách
for mau_nuoc in danh_sach_do_duc:
  if mau_nuoc > 5:
        print("Mẫu nước có độ đục " + str(mau_nuoc) + " là NƯỚC BẨN. Xả bỏ!")
  else:
        print("Mẫu nước có độ đục " + str(mau_nuoc) + " là NƯỚC SẠCH. Cho qua!")
```
